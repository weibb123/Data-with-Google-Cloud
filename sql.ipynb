```python
instance_connection_name = f"{project_id}:{region}:{instance_name}"
print(f"Your instance connection name is: {instance_connection_name}")
DB_USER = "wei"
DB_PASS = 'my_password'
DB_NAME = "hw5" 
```

```python
from google.cloud.sql.connector import Connector
import sqlalchemy
from sqlalchemy import text

# initialize connector object
connector = Connector()

# function to return database connection object
def getconn():
    conn = connector.connect(
        instance_connection_name,
        'pymysql',
        user=DB_USER,
        password=DB_PASS,
        db=DB_NAME
    )
    return conn

# create connection pool with 'creator' argument to our connection object
pool = sqlalchemy.create_engine(
    'mysql+pymysql://',
    creator=getconn,
)
```

```python
with pool.connect() as db_conn:
   db_conn.execute(text(
    'DROP TABLE IF EXISTS IP_TABLE'
    ))
   db_conn.commit()
```

```python
with pool.connect() as db_conn:
    db_conn.execute(text(
        "CREATE TABLE IF NOT EXISTS IP_TABLE "
        "(id SERIAL NOT NULL, country VARCHAR(255) NOT NULL, gender VARCHAR(255) NOT NULL, "
        "age VARCHAR(255) NOT NULL, income VARCHAR(255) NOT NULL, "
        "client_ip VARCHAR(255), time_of TIME, requested_file VARCHAR(255) NOT NULL, "
        "PRIMARY KEY (id));"
        )
    )
```
